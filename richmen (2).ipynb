```python
# fill in 0 to all the rows with nan in age column
df['Age'] = df['Age'].fillna(0)
df
```

```python
# removing allthe rows with nan 
not_null_roll = df.notnull().all(axis=1)
data = df[not_null_roll]
data
```

```python
# checking for null values
data.isnull().sum()
```

```python
# converting 'Age' column to integer
data['Age'] = data['Age'].astype(int)
data
```

```python
#removing the dollar and billion sign on 'net wort'
data['Net Worth'] = data['Net Worth'].str.replace('[$,B]', '', regex=True).astype(float)
```

```python
data
```

```python
data.describe()
```

```python
result = data.groupby('Country')[['Net Worth']].mean().sort_values (by= 'Net Worth',ascending=False).head(10)
```

```python
result
```

```python
import matplotlib.pyplot as plt
```

```python
result.plot(kind='bar', y='Net Worth', figsize=(10, 6))
plt.title('Top 10 Countries with Highest Mean of Net Worth')
plt.xlabel('Country')
plt.ylabel('Mean')
plt.xticks(rotation=45)
plt.show()
```
